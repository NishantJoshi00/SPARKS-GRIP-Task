## Imported Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df['Species'].head()
```

## Data Visualisation

### For unsupervised learning expected outputs are not present, but here we can use them to test the accuracy of the self trained model at the end

```python
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.scatterplot(x = 'SepalLengthCm', y = 'PetalLengthCm', data = df, hue = 'Species', palette = 'crest')
```

## Training the model

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(df.drop(['Species'],axis=1))
```

```python
km.labels_
```

## Visualising and comparing expected and actual result

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
sns.scatterplot(ax = ax1,x = 'SepalLengthCm', y = 'PetalLengthCm', data = df, hue = 'Species', palette = 'rocket',legend=False)
ax2.set_title("Original")
sns.scatterplot(ax = ax2, x = 'SepalLengthCm', y = 'PetalLengthCm', data = df, hue = km.labels_, palette = 'crest',legend=False)
```

### As one can see, the result of the unsupervied learning model is very close to that of the expected result
